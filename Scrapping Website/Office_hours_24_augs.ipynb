```python
# Let's use a dictioonary to store the data/information we scrapped
# Let's use functions to encapsulate our work 
def pull_info(product_el): 
    """Return a dictionary of title, description, and price contained in HTML element."""
    title = product_el.select_one('a').text
    price = product_el.select_one('h4.price').text.replace('$', '')
    description = product_el.select_one('p').text
    
    return { 
        'title': title,
        'price': price,
        'description' : description
    }
```

```python
# I want to encapsulate the process for going from 
# URL -> list of dictionaries
# Break it down into
# URL -> soup object
# soup -> list of dictionaries

import json 

def create_soup(url): 
    """return a Beatiful Soup Object from a given url"""
    response = requests.get(url)
    return BeautifulSoup(response.text);

def scrape_products(soup):
    """return a list of dictionaries containing the scrapped
    data for each product"""
    el_products = soup.select('div.caption')
    data = [pull_info(el) for el in el_products]
    return data

def dum_data(data): 
    """Dump list of dictionaries to disk as a JSON"""
    with open('output.json', 'w') as f: 
        json.dump(data, f, indent=2)

def main(url):
    soup = create_soup(url)
    data = scrape_products(soup)
    dum_data(data)
    return data


main('https://webscraper.io/test-sites/e-commerce/static/computers/laptops')
```

```python
# I want to encapsulate the process for going from 
# URL -> list of dictionaries
# Break it down into
# URL -> soup object
# soup -> list of dictionaries

import json 

def create_soup(url_base, page): 
    """return a Beatiful Soup Object from a given url"""
    params = {'page': page }
    response = requests.get(url_base, params=params)
    return BeautifulSoup(response.text);

def scrape_products(soup):
    """return a list of dictionaries containing the scrapped
    data for each product"""
    el_products = soup.select('div.caption')
    data = [pull_info(el) for el in el_products]
    return data

def dum_data(data): 
    """Dump list of dictionaries to disk as a JSON"""
    with open('output.json', 'w') as f: 
        json.dump(data, f, indent=2)

def scrape_page(url_base, page):
    """Scrape data for one page"""
    soup = create_soup(url_base, page)
    data = scrape_products(soup)
    return data

def scrape_page_all(url_base, page_limit = 5): 
    return [scrape_page(url_base, page) for page in range(1, page_limit + 1)]
        
def main(url_base):
    data_all = scrape_page_all(url_base, page_limit=20)
    dum_data(data_all)
    return data_all


main('https://webscraper.io/test-sites/e-commerce/static/computers/laptops')
```
